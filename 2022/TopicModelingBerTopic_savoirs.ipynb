```python
import os
import pandas as pd

lst = []
def load_files(lst, folder):
  for file_name in os.listdir(folder):
    if file_name.endswith(".txt"):
      f_txt = open(os.path.join(folder, file_name), "r")
      data_list = f_txt.read().split()[10:5000]
      data = ' '.join(data_list)
      #print(data)
      #print("")
      lst.append([file_name, data])
  return lst

lst = load_files(lst, "/content/drive/MyDrive/Colab Notebooks/data/savoirs_txt")
lst = load_files(lst, "/content/drive/MyDrive/Colab Notebooks/data/articles_cairn_ne_pas_partager/assr")
lst = load_files(lst, "/content/drive/MyDrive/Colab Notebooks/data/articles_cairn_ne_pas_partager/lhomme")
lst = load_files(lst, "/content/drive/MyDrive/Colab Notebooks/data/articles_cairn_ne_pas_partager/etudesafricaines")
lst = load_files(lst, "/content/drive/MyDrive/Colab Notebooks/data/articles_cairn_ne_pas_partager/etudesrurales")

df = pd.DataFrame(lst, columns =['article_file', 'text_article'])
df
```

```python
# create model 
 
model = BERTopic(language="multilingual", verbose=True)
 
#convert to list 
docs = df.text_article.to_list()
#docs 

topics, probabilities = model.fit_transform(docs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
stop_words = ["de","des","le","les", "la", "en", "de","et", "une", "que", "dans", "du", "un", "il", "elle", "qui", "pour", "pas", "ce", "qu", "par", "au", "ne", "ou", "sur", "cette", "se", "avec", "mais", "ces", "est", "sont", "plus", "aux", "on", "sa", "ses", "comme", "était"]
vectorizer_model = CountVectorizer(stop_words=stop_words, ngram_range=(1, 2))
model.update_topics(docs, topics, vectorizer_model=vectorizer_model)
```
