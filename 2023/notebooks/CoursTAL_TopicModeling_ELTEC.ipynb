```python
txt_folder_fra = "/content/ELTeC-fra/plain1"
txt_folder_fra_ext = "/content/ELTeC-fra-ext2/plain"
```

```python
import glob

def fetch_eltec_fr(folder_name):
  docs = []
  for file in glob.glob(folder_name+"/*.txt"):
    with open(file, encoding="utf-8") as f:
      contents = f.read()
      #print(contents)
      print(file)
      docs.append(contents)
  return docs
```

```python
docs_eltec_fr = fetch_eltec_fr(txt_folder_fra)
```

```python
docs_eltec_fr_ext = fetch_eltec_fr(txt_folder_fra_ext)
```

```python
def eltec_fr_subset(docs, nb_sent):
  sub_docs = []
  for doc in docs:
    doc_lines = doc.split("\n")
    subcontent = '\n'.join(doc_lines[1:nb_sent])
    sub_docs.append(subcontent.replace(u'\xa0', u' '))
  return sub_docs
```

```python
docs_eltec_fr_subset = eltec_fr_subset(docs_eltec_fr + docs_eltec_fr_ext, 50)
```

```python
# Import dependencies
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
import spacy
import pyLDAvis
import pyLDAvis.gensim_models
from sklearn.datasets import fetch_20newsgroups
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
def lemmatize(docs, allowed_postags = ["NOUN", "ADJ", "VERB", "ADV"]):
  '''
  Performs lemmization of input documents.
  Args:
    - docs: list of strings with input documents
    - allowed_postags: list of accepted Part of Speech (POS) types
  Output:
    - list of strings with lemmatized input
  '''
  nlp = spacy.load("fr_core_news_sm", disable = ["parser", "ner"])
  nlp.max_length = 3000000 #this is possible as we're not using parser or ner.
  lemmatized_docs = []
  for doc in docs:
    doc = nlp(doc)
    tokens = []
    for token in doc:
      if token.pos_ in allowed_postags:
        tokens.append(token.lemma_)
    lemmatized_docs.append(" ".join(tokens))
  return (lemmatized_docs)
```

```python
lemmatized_docs_eltec_fr = lemmatize(docs_eltec_fr_subset)
```

```python
lemmatized_docs_eltec_fr[0]
```

```python
def tokenize(docs):
  '''
  Performs tokenization of input documents.
  Args:
    - docs: list of strings with input documents
  Output:
    - list of strings with tokenized input
  '''
  tokenized_docs = []
  for doc in docs:
    tokens = gensim.utils.simple_preprocess(doc, deacc=True)
    tokenized_docs.append(tokens)
  return (tokenized_docs)
```

```python
tokenized_docs_eltec_fr = tokenize(lemmatized_docs_eltec_fr)
```

```python
tokenized_docs_eltec_fr[0]
```

```python
# Mapping from word IDs to words
id2word = corpora.Dictionary(tokenized_docs_eltec_fr) 

# Prepare Document-Term Matrix
corpus = []
for doc in tokenized_docs_eltec_fr: 
    corpus.append(id2word.doc2bow(doc))

# Fit LDA model: 
topic_model = gensim.models.ldamodel.LdaModel(
    corpus = corpus,      # Document-Term Matrix
    id2word = id2word,    # Map word IDs to words
    num_topics = 50,      # Number of latent topics to extract
    random_state = 100,
    passes = 1000,         # N° of passes through the corpus during training
    )
```

```python
print(topic_model.print_topics())
```

```python
# Visualize with pyLDAvis: See [2] for more details
pyLDAvis.enable_notebook()
visualization = pyLDAvis.gensim_models.prepare(
    topic_model, 
    corpus,
    id2word, 
    mds = "mmds", 
    R = 50)
```

```python
visualization
```

```python
from gensim.models import CoherenceModel

coherence_model_lda = CoherenceModel(model=topic_model, texts=tokenized_docs_eltec_fr, dictionary=id2word)
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```

Pour un bon modèle LDA la cohérence doit être comprise entre 0.4 et 0.7 au-delà et en dessous le modèle est très probablement erroné.


The approach to finding the optimal number of topics is to build many LDA models with different values of a number of topics (k) and pick the one that gives the highest coherence value.
