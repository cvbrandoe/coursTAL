```python
import json, glob

directory_json = 'coursTAL/2023/EltecFrJsonTagtog_annot_revis'
directory_tr = "coursTAL/2023/EltecFRTxtTrain" 

#manual_data = {'classes' : ['PER', "LOC", "ORG", "MISC"], 'annotations' : []}
manual_data = {'classes' : ['PER', "LOC", "ORG", "OTHER", "WORK", "DEMO", "ROLE","EVENT"], 'annotations' : []}

#classes modèle spacy habituel : PER, LOC, ORG, MISC,
#classes annot manuelle : PERS, LOC, ORG, OTHER, WORK, DEMO, ROLE

def get_class_id_inv(label):
  choices = {'e_1': 'PER', 'e_2': 'LOC', 'e_3': 'ORG','e_4': 'OTHER','e_5': 'WORK','e_6': 'DEMO','e_7': 'ROLE','e_8': 'EVENT'}
  return choices.get(label, None)

for filename in glob.glob(directory_json+'/*.json'):
  filen = filename.split("/")[-1]
  print(filen)
  with open(filename, "r", encoding="utf_8") as f:
    data = f.read()  
    js_corrected_manual_data = json.loads(data)
   
    temp_dict = {}
    fin = open(directory_tr + "/tr_FRA" + filen.split("-FRA")[-1].replace(".ann.json","").replace("Viel_Castel","Viel-Castel"), "r")
    content_text = fin.read() #.replace(u'\xa0', u' ')
    temp_dict['text'] = content_text
    #print(content_text)
    fin.close()
    temp_dict['entities'] = []
    
    for example in js_corrected_manual_data['entities']:  
      #print(example)
      start = example['offsets'][0]['start']
      end = int(example['offsets'][0]['start']) + len(example['offsets'][0]['text'])
      label = get_class_id_inv(example['classId'])
      temp_dict['entities'].append((start, end, label))
    manual_data['annotations'].append(temp_dict)
```

```python
import random

annot_data = manual_data['annotations']
print(len(annot_data))

# Mélange des données
random.shuffle(annot_data) 

# Calcul de l'index de séparation pour les données de test
test_size = int(len(annot_data) / 2)
test_data = annot_data[:test_size]
train_data = annot_data[test_size:]
```

```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm
from spacy.util import filter_spans

def create_bin(annots, lab, folderout):
  nlp = spacy.blank("fr") # load a new spacy model
  doc_bin = DocBin() # create a DocBin object
  for example  in tqdm(annots): 
      #print(example)
      text = example['text']
      labels = example['entities']
      doc = nlp.make_doc(text) 
      ents = []
      for start, end, label in labels:
          span = doc.char_span(start, end, label=label)
          if span is None:
              print("Skipping entity")
          else:
              ents.append(span)
      filtered_ents = filter_spans(ents)
      doc.ents = filtered_ents 
      doc_bin.add(doc)
  doc_bin.to_disk(folderout+"/"+lab+"_data.spacy") # save the docbin object
```

```python
create_bin(train_data, "train", ".")
create_bin(test_data, "test", ".")
```

Cross fold validation

```python
import json
import os
import random

# Fraction des données à utiliser pour l'ensemble de test
test_frac = 0.2
test_size = int(len(annot_data) * test_frac)
print(len(annot_data))
# Nombre de plis pour la validation croisée
n_folds = 5

# Parcours des plis pour la validation croisée
for i in range(n_folds):
    # Chemin du répertoire pour ce pli
    fold_path = os.path.join('folds', f'fold_{i}')
    print(fold_path)
    # Création du répertoire
    os.makedirs(fold_path, exist_ok=True)

    start = i * test_size
    end = (i + 1) * test_size
    print("start - end: "+str(start)+" - " + str(end))
    
    test_data = annot_data[start:end]
    print("test:"+str(len(test_data)))
    create_bin(test_data, "test", fold_path)

    train_data = annot_data[:start] + annot_data[end:]
    print("train:"+str(len(train_data)))
    create_bin(train_data, "train", fold_path)
    
```
