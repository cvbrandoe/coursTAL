```python
import json, glob

directory_json = 'coursTAL/2023/EltecFrJsonTagtog_annot_revis'
directory_tr = "coursTAL/2023/EltecFRTxtTrain" 

#training_data = {'classes' : ['PER', "LOC", "ORG", "MISC"], 'annotations' : []}
training_data = {'classes' : ['PER', "LOC", "ORG", "OTHER", "WORK", "DEMO", "ROLE","EVENT"], 'annotations' : []}

#classes modèle spacy habituel : PER, LOC, ORG, MISC,
#classes annot manuelle : PERS, LOC, ORG, OTHER, WORK, DEMO, ROLE

def get_class_id_inv(label):
  choices = {'e_1': 'PER', 'e_2': 'LOC', 'e_3': 'ORG','e_4': 'OTHER','e_5': 'WORK','e_6': 'DEMO','e_7': 'ROLE','e_8': 'EVENT'}
  return choices.get(label, None)

for filename in glob.glob(directory_json+'/*.json'):
  filen = filename.split("/")[-1]
  print(filen)
  with open(filename, "r", encoding="utf_8") as f:
    data = f.read()  
    #print("Data type before reconstruction : ", type(data))
    # reconstructing the data as a dictionary
    js_corrected_training_data = json.loads(data)
    print("Data type after reconstruction : ", type(js_corrected_training_data))
    print(js_corrected_training_data)
    temp_dict = {}
    fin = open(directory_tr + "/tr_FRA" + filen.split("-FRA")[-1].replace(".ann.json","").replace("Viel_Castel","Viel-Castel"), "r")
    temp_dict['text'] = fin.read()
    fin.close()
    temp_dict['entities'] = []
    for example in js_corrected_training_data['entities']:  
      #print(example)
      start = example['offsets'][0]['start']
      end = int(example['offsets'][0]['start']) + len(example['offsets'][0]['text'])
      label = get_class_id_inv(example['classId'])
      temp_dict['entities'].append((start, end, label))
      #print(temp_dict['entities'])
    training_data['annotations'].append(temp_dict)
```

```python
import json, glob

import csv

def get_class_id_inv(label):
  choices = {'e_1': 'PER', 'e_2': 'LOC', 'e_3': 'ORG','e_4': 'OTHER','e_5': 'WORK','e_6': 'DEMO','e_7': 'ROLE','e_8': 'EVENT'}
  return choices.get(label, None)

count = {}
keyList = ['PER', 'LOC', 'ORG', 'OTHER', 'WORK', 'DEMO', 'ROLE', 'EVENT']
for i in keyList:
  count[i] = 0
  
with open('count_cat_eltecfr.csv', 'w') as f:
  f.write("file,")
  for key in keyList:    
    f.write("%s,"%(key)) 
  f.write("\n")
f.close()

for filename in glob.glob(directory_json+'/*.json'):
  count = {}
  keyList = ['PER', 'LOC', 'ORG', 'OTHER', 'WORK', 'DEMO', 'ROLE', 'EVENT']
  for i in keyList:
    count[i] = 0
  filen = filename.split("/")[-1].replace(".ann.json","").split("-")[1]
  print(filen)
  with open(filename, "r", encoding="utf_8") as f:
    data = f.read()  
    js_corrected_training_data = json.loads(data)
    for example in js_corrected_training_data['entities']:  
      label = get_class_id_inv(example['classId'])
      count[label] = count[label] + 1  
    print(count)
    with open('count_cat_eltecfr.csv', 'a') as f2:
      f2.write("%s,"%(filen))
      for key in count.keys():
        f2.write("%s,"%(count[key])) 
      f2.write("\n")
    f2.close()
```

```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm

nlp = spacy.blank("fr") # load a new spacy model
doc_bin = DocBin() # create a DocBin object

from spacy.util import filter_spans

for training_example  in tqdm(training_data['annotations']): 
    print(training_example)
    text = training_example['text']
    labels = training_example['entities']
    doc = nlp.make_doc(text) 
    ents = []
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label)
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents 
    doc_bin.add(doc)

doc_bin.to_disk("training_data.spacy") # save the docbin object
```
